```python
# Importação das bibliotecas
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# Importação dos dados file *.xlsx
df = pd.read_csv('C:/Users/profc/OneDrive/Documentos/COMUSA.csv', sep=';', encoding='utf8', decimal=',', usecols = [0,1,10,11,12,13,14,15,16])

# Excluir os campos de 'Total a Pagar (R$)' até 'Valor Água (R$)'
```

```python
# Cálculo da média de todos os indicadores
cor = df['Cor (UC)']
coliformes = df['% Coliformes Totais'].mean()
turbidez = df['Turbidez (NTU)'].mean()
dioxido_cloro = df['Dióxido Cloro (mg/l)'].mean()
cloro_livre = df['Cloro Livre (mg/l)'].mean()
fluor = df['Flúor (mg/l)'].mean()
pH = df['pH'].mean()
```

```python
# Gráfico do pH
my_dpi = 200
plt.figure(figsize=(1100/my_dpi, 400/my_dpi), dpi = my_dpi)
 
# Plot do pH
plt.plot(df['Mês/Ano Referência'], df['pH'], marker='', color='black', linewidth=1.5)

# Add a title to say which style it is
plt.title('COMUSA: pH', fontsize=12, fontweight=0, color='black', loc='center')
plt.xlabel('Período', fontsize=6)
plt.ylabel('pH', fontsize=6, rotation=0)
plt.xticks(fontsize=4, rotation=70)
plt.yticks(fontsize=4)

# Add limits to Y axis
yinf = 5
ysup = 11
plt.ylim(yinf, ysup)

# Add lines for pH Parameters
plt.axhline(9.5, color ='b', linestyle = "dashdot")
plt.axhline(6, color ='b', linestyle = "dashdot")

plt.show()
```
